```python
# importing 'requests' library and 'BeautifulSoup' function from 'bs4' library 
import requests
from bs4 import BeautifulSoup as bs
```

```python
# assigning web-link/web-address to 'url' to access the website from where the data are to be downloaded
filepath = "https://en.wikipedia.org/wiki/List_of_counties_in_Maryland#List_of_counties"
```

```python
# getting data from the website
file = requests.get(filepath)
```

```python
# checking the status of the accessed'file'-- HTTP Status Code means 'OK'
file.status_code
```

```python
# checking the accessed text in 'file'
file.text
```

```python
# saving accessed file content as string
file_str = file.text
print(file_str[:150])
```

```python
# parse the html content in 'file_str' using 'html.parser' function in BeautifulSoup or 'bs' 
# and save it in variable 'sp'
sp = bs(file_str, "html.parser")
```

```python
#find all the <table> tags
all_tables=sp.find_all('table')
all_tables
```

```python
# get the <table> tag that contains the data we want to scrape
data_table=sp.find('table', class_='wikitable sortable')
print(data_table)
```

```python
#set empty lists to hold data of each column
CN=[]
FC=[]
CS=[]
EY=[]
OR=[]
ET=[]
PP=[]
AR=[]
```

```python
# find all <tr> tags in the table and go through each one (row)
# tr table row tag
for row in data_table.findAll("tr"):
    
    #get all the <th> and <td> tags for each <tr> tag
    county = row.findAll('th')
    cells = row.findAll('td')
    
    #if there are 10 <td> tags, 10 cells in a row
    if len(cells)==10:
        CN.append(county[0].find(text=True))
        FC.append(cells[0].find(text=True))
        CS.append(cells[1].find(text=True))
        EY.append(cells[2].find(text=True))
        OR.append(cells[3].find(text=True))
        ET.append(cells[4].find(text=True))
        PP.append(cells[7].find(text=True))
        AR.append(cells[8].find(text=True))
```

```python
# getting the length of 'CN' list
print(len(CN))
```

```python
#importing pandas as pd to convert the lists to data frame
import pandas as pd

dataframe=pd.DataFrame(CN, columns=['County Name']) #turn list CN into dataframe
dataframe.head()
```

```python
#add other lists as new columns in my new dataframe
dataframe['FIPS Code'] = FC
dataframe['County Seat'] = CS
dataframe['Established (Year)'] = EY
dataframe['Origin'] = OR
dataframe['Etymology'] = ET
dataframe['Population'] = PP
dataframe['Area'] = AR

#show first 5 rows of created dataframe
dataframe.head()
```

```python
#export scraped data to a csv file
dataframe.to_csv("CountiesInMaryland.csv")
```
